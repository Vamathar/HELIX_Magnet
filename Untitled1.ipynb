```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
```

```python
%matplotlib notebook
```

```python
measured_filename = "measured_field.csv"
calculated_filename = "calculated_field.csv"
measured_df = pd.read_csv(measured_filename,header=None)
calculated_df = pd.read_csv(calculated_filename,header=None)
```

```python
fig = plt.figure(figsize=(8,7))
ax = fig.add_subplot(111,projection='3d')
ax.set_xlim(measured_df[0].min(),measured_df[0].max())
ax.set_ylim(measured_df[1].min(),measured_df[1].max())
ax.set_zlim(measured_df[2].min(),measured_df[2].max())
x,y,z = measured_df[0],measured_df[1],measured_df[2]
u,v,w = measured_df[3],measured_df[4],measured_df[5]
ax.quiver(x,y,z,u,v,w,length=100, pivot='middle')
#ax.scatter(x,y,z)
plt.show()
```
